### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
import matplotlib.pyplot as plt

top5 = df.sort_values("IMDB_Rating", ascending=False).head(5)

imdb = top5["IMDB_Rating"] * 10
meta = top5["Meta_score"]

x = range(len(top5))

plt.figure(figsize=(8, 5))

plt.bar(x, imdb, width=0.4, label="IMDB x10")
plt.bar([i + 0.4 for i in x], meta, width=0.4, label="Meta Score")

plt.xticks([i + 0.2 for i in x], top5["Series_Title"], rotation=45)

plt.xlabel("Movies")
plt.ylabel("Score")
plt.title("Top 5 Movies Comparison")
plt.legend()

plt.tight_layout()
plt.show()
```
